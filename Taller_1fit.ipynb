````python
def rigthRotatebyOne(cubo):
  par = []
  impar = []
  for i in range(len(cubo)):
    if((i+1)%2 == 0):
      par.append(cubo[i])
    if((i+1)%2 != 0):
      impar.append(cubo[i])
  cubo= []
  cubo = par + impar
  return cubo
````

````python
def merge_two_dicts(x, y):
    z = x.copy()   # start with keys and values of x
    z.update(y)    # modifies z with keys and values of y
    return z
````

````python
def transition(state,action):
  for a in action:
    state.cube[a]=action[a]
  state.depth=state.depth+1


````

````python
def get_valid_actions(state):
  actions=[]
  
  actions.append(turnRigth(0))
  actions.append(turnRigth(1))
  actions.append(turnRigth(5))
  return actions
````

````python
def is_final_state(state):
  for i in state.cube:
    if(not(i.count(i[0])==len(i))):
      return False
 
  return True
````

````python
from copy import deepcopy

def bfs(initial_state):
  queue = [deepcopy(initial_state)]
  iters = 0
  while len(queue)>0:
    iters += 1
    print(iters)
    state = queue.pop(0) #retorna el primer elemento y lo elimina 
    
    if is_final_state(state): return state,iters #se encontr√≥ una soluci√≥n!

    actions = get_valid_actions(state)

    for action in actions:
      new_state = deepcopy(state) #ojo que debemos copiar el estado antes de modificarlo!
      transition(new_state, action)
      queue.append(new_state) #agrega al final
````
